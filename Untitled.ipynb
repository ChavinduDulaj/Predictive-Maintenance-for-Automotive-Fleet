```python
import pandas as pd
import random
import datetime

# Function to generate synthetic data
def generate_data(n):
    data = {
        'Vehicle_Model': [random.choice(['Car', 'SUV','Van' ,'Truck','Bus','Motorcycle']) for _ in range(n)],
        'Mileage': [random.randint(30000, 80000) for _ in range(n)],
        'Maintenance_History': [random.choice(['Good', 'Average', 'Poor']) for _ in range(n)],
        'Reported_Issues': [random.randint(0, 5) for _ in range(n)],
        'Need_Maintenance': [random.randint(0, 1) for _ in range(n)],
        'Vehicle_Age': [random.randint(1, 10) for _ in range(n)],
        'Fuel_Type': [random.choice(['Diesel', 'Petrol', 'Electric']) for _ in range(n)],
        'Transmission_Type': [random.choice(['Automatic', 'Manual']) for _ in range(n)],
        'Engine_Size': [random.choice([800,1000, 1500, 2000, 2500]) for _ in range(n)],
        'Odometer_Reading': [random.randint(1000, 150000) for _ in range(n)],
        'Last_Service_Date': [(datetime.datetime.now() - datetime.timedelta(days=random.randint(30, 365))).strftime('%Y-%m-%d') for _ in range(n)],
        'Warranty_Expiry_Date': [(datetime.datetime.now() + datetime.timedelta(days=random.randint(30, 365*2))).strftime('%Y-%m-%d') for _ in range(n)],
        'Owner_Type': [random.choice(['First', 'Second', 'Third']) for _ in range(n)],
        'Insurance_Premium': [random.randint(5000, 30000) for _ in range(n)],
        'Service_History': [random.randint(1, 10) for _ in range(n)],
        'Accident_History': [random.randint(0, 3) for _ in range(n)],
        'Fuel_Efficiency': [random.uniform(10, 20) for _ in range(n)],
        'Tire_Condition': [random.choice(['New', 'Good', 'Worn Out']) for _ in range(n)],
        'Brake_Condition': [random.choice(['New', 'Good', 'Worn Out']) for _ in range(n)],
        'Battery_Status': [random.choice(['New', 'Good', 'Weak']) for _ in range(n)]
    }
    return data

# Generate synthetic data
n = 50000
data = generate_data(n)

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('vehicle_maintenance_data.csv', index=False)

# Display the DataFrame
print(df.head())
```
